## Coding Section

```python
# Importing libraries/Toolkits
# Import Matplotlib
import matplotlib.pyplot as plt
# Import 3D Visualization library
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Given X,Y,and Z Coordinates
x = [-0.41, 0.57, 0.07, 0.00, -0.29, -0.32,-0.50,-0.23, -0.23]
y = [4.12, 7.71, 2.36, 9.10, 13.35, 8.13, 7.19, 13.25,13.43]
z = [2.06, 0.84, 1.56, 2.07, 2.36, 1.72, 0.66, 1.25,1.38]
```

```python
# 2D Visualization
# Creating a figure
fig = plt.figure()
# Adding the subplot
ax = fig.add_subplot(1, 1, 1)
# Creating a scatter plot
ax.scatter(x, y)
plt.show()
```

```python
# 3D Visualization
# Creating a figure
fig_3d = plt.figure()
# Adding the subplot
ax_3d = fig_3d.add_subplot(111, projection='3d')
# Creating a scatter plot
ax_3d.scatter(x, y, z)
plt.show()
```

```python
# Adding Color to 2D Grid/Points
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
# Changing color of plot background
ax.set_facecolor('black')
# Creating a scatter plot with black background
ax.scatter(x, y, color='white')
plt.show()
```

```python
# Creating a figure
fig_3d = plt.figure()
# Adding your subplot
ax_3d = fig_3d.add_subplot(111, projection='3d')
# Change color of plot background
ax_3d.set_facecolor('black')
# Create a scatter plot with black background
ax_3d.scatter(x, y, color='white')
plt.show()
```

```python
# Adding Color to 2D Grid/Points
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
# Changing color of plot background
ax.set_facecolor('black')
# Creating a scatter plot with black background
ax.scatter(x, y, color='white')
# Titling 2D
plt.title("Orion Constellation (2D View from Earth)")
plt.show()
```

```python
# Creating a figure
fig_3d = plt.figure()
# Adding your subplot
ax_3d = fig_3d.add_subplot(111, projection='3d')
# Change color of plot background
ax_3d.set_facecolor('black')
# Create a scatter plot with black background
ax_3d.scatter(x, y, color='white')
# Titling 3D
plt.title("Orion Constellation (3D View from Earth)")
plt.show()
```
